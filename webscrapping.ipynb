```python
# Question - 1
page1=requests.get("https://www.wikipedia.org/")
wiki_soup=BeautifulSoup(page1.content,"html.parser")
header_tags = wiki_soup.find_all(["h1", "h2", "h3", "h4", "h5", "h6"])
for h in header_tags:
    print(h)
    print(" ")
```

```python
# Question - 2
movie_names = []
release_years = []
ratings = []
page2=requests.get("https://www.imdb.com/list/ls091520106/")
movie_soup=BeautifulSoup(page2.content)
movie_list = movie_soup.find_all("h3", class_="lister-item-header")
for c in movie_list:
    name = c.contents[3].text
    movie_names.append(name)
for y in movie_soup.find_all("span",class_="lister-item-year text-muted unbold"):
    release_years.append(y.text.replace('(', '').replace(')', ''))
for r in movie_soup.find_all("div",class_="ipl-rating-star small"):
    ratings.append(r.contents[3].text)
    
    
df=pd.DataFrame({"Name":movie_names," Rating":ratings,"Year of release":release_years})
df
```

```python
# Question - 3
titles=[]
location=[]
cuisine_array=[]
rating=[]
images=[]
page3=requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
soup=BeautifulSoup(page3.content)
cuisine=soup.find_all("span",class_="double-line-ellipsis")
for c in cuisine:
    cuisine_string=''
    for item in c.contents[2:]:
        cuisine_string = cuisine_string + item.text    
    cuisine_array.append(cuisine_string)
for i in soup.find_all("div",class_="restnt-info cursor"):
    titles.append(i.text)
for l in soup.find_all("div",class_="restnt-loc ellipsis"):
    location.append(l.text)
for r in soup.find_all("div",class_="restnt-rating rating-4"):
    rating.append(r.text)
for i in soup.find_all("img",class_="no-img"):
    images.append(i["data-src"])
df=pd.DataFrame({"Title":titles,"Location":location,"Cuisine":cuisine_array,"Rating":rating,"Images":images})
df
```

```python
# Question - 4
president_names=[]
terms = []

main_page=requests.get("https://presidentofindia.nic.in/former-presidents")
main_soup=BeautifulSoup(main_page.content)
for n in main_soup.find_all("div",class_="desc-sec"):
    name_text = n.find("h3").text
    president_names.append(name_text)
all_presidents = main_soup.findAll('div', class_='president-listing')
for p in all_presidents:
    page_link = p.find('a').get('href')
    try:
        president_page = requests.get(page_link)
        president_page_soup = BeautifulSoup(president_page.content)
        term_text = president_page_soup.find("span", class_="bot-txt")
        dates = term_text.text.replace("Term of Office:","")
        terms.append(dates)
    except Exception as err:
        terms.append('')
        print(f'Error occured in fetching president page')


df=pd.DataFrame({"Name":president_names,"Term_of_Office":terms})
df
```
