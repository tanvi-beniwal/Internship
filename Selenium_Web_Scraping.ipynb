```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
```

# Q 1


```python
driver=webdriver.Chrome()
driver.get("https://www.naukri.com/")

designation=driver.find_element(By.CLASS_NAME,"suggestor-input ")
designation.send_keys("Data Scientist")

location=driver.find_element(By.XPATH,"/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input")
location.send_keys("Delhi/NCR")

search=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
time.sleep(10)
salary_filter = driver.find_element(By.XPATH, "//label[@for='chk-3-6 Lakhs-ctcFilter-']")
salary_filter.click()
time.sleep(25)

job_title=[]
job_location=[]
company_name=[]
experience_required=[]
salary_mentioned=[]

title_tags=driver.find_elements(By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div/a')
for i in title_tags:
    title=i.text
    job_title.append(title)

location_tags=driver.find_elements(By.XPATH,'//span[@class="loc-wrap ver-line"]/span/span')
for i in location_tags:
    location=i.text
    job_location.append(location)

company_tags=driver.find_elements(By.XPATH,'//div[@class=" row2"]/span/a[1]')
for i in company_tags:
    company=i.text
    company_name.append(company)
    
experience_tags=driver.find_elements(By.XPATH,'//span[@class="expwdth"]')
for i in experience_tags:
    experience=i.text
    experience_required.append(experience)

salary_tags=driver.find_elements(By.XPATH,'//span[@class="ni-job-tuple-icon ni-job-tuple-icon-srp-rupee sal"]/span')
for i in salary_tags:
    salary=i.text
    salary_mentioned.append(salary)

job_title = job_title[0:10]
job_location = job_location[0:10]
company_name = company_name[0:10]
experience_required = experience_required[0:10]
salary_mentioned = salary_mentioned[0:10]


df=pd.DataFrame({"Title":job_title,'Location':job_location,'Company':company_name,'Experience':experience_required,'Salary':salary_mentioned})
df
```

# Q3

```python
Rating_title=[]
Review_summary=[]
Full_review=[]

driver=webdriver.Chrome()


def scrape_data():
    Rating_tags=driver.find_elements(By.XPATH,'//div[@class="XQDdHH Ga3i8K"]')
    for i in Rating_tags:
        Rating=i.text
        Rating_title.append(Rating)

    Review_tags=driver.find_elements(By.XPATH,'//p[@class="z9E0IG"]')
    for i in Review_tags:
        Review=i.text
        Review_summary.append(Review)

    Fullreview_tags=driver.find_elements(By.XPATH,'//div[@class="ZmyHeo"]/div/div')
    for i in Fullreview_tags:
        Fullreview=i.text
        Full_review.append(Fullreview)


for count in range(1,11):
    url = "https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART&page=" + str(count)
    driver.get(url)
    time.sleep(3)
    scrape_data()


df=pd.DataFrame({"Title":Rating_title,'Review':Review_summary,'Full_review':Full_review})
df
```

# Q4

```python
Brand_title=[]
Product_Description=[]
Price_title=[]


def scrape_data():
    Brand_tags=driver.find_elements(By.XPATH,'//div[@class="syl9yP"]')
    for i in Brand_tags:
        Brand=i.text
        Brand_title.append(Brand)

    Product_tags=driver.find_elements(By.XPATH,'//a[@class="WKTcLC"]')
    for i in Product_tags:
        Product=i.text
        Product_Description.append(Product)

    Price_tags=driver.find_elements(By.XPATH,'//a[@class="+tlBoD"]/div/div')
    for i in Price_tags:
        Price=i.text
        Price_title.append(Price)
        
driver=webdriver.Chrome()
driver.get("https://www.flipkart.com")
Product=driver.find_element(By.CLASS_NAME,"Pke_EE")
Product.send_keys("sneakers")

search=driver.find_element(By.CLASS_NAME,"_2iLD__")
search.click()
time.sleep(5)
scrape_data()
 
for count in range(1,4):
    next_buttons = driver.find_elements(By.CLASS_NAME, "_9QVEpD")
    if(len(next_buttons) > 1):
        next_button = next_buttons[1]
    else:
        next_button = next_buttons[0]
    next_button.click()
    time.sleep(5)
    scrape_data()

Brand_title = Brand_title[1:101]
Product_Description=Product_Description[1:101]
Price_title=Price_title[1:101]

df=pd.DataFrame({"Brand":Brand_title,'Description':Product_Description,'Price':Price_title})
df
```

# Q6

```python
Quote_title=[]
Author_title=[]
Type_of_quotes_title=[]

def scrape_data():
    Quote_tags=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in Quote_tags:
        Quote=i.text
        Quote_title.append(Quote)

    Author_tags=driver.find_elements(By.XPATH,'//div[@class="author"]/a')
    for i in Author_tags:
        Author=i.text
        Author_title.append(Author)

    Type_of_quotes_tags=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in Type_of_quotes_tags:
        Type_of_quotes=i.text
        Type_of_quotes_title.append(Type_of_quotes)

        
driver=webdriver.Chrome()
driver.get("https://www.azquotes.com/")

top_quote_elem = driver.find_element(By.XPATH, "//a[normalize-space()='Top Quotes']")
top_quote_elem.click()
time.sleep(10)
scrape_data()

for count in range(1,10):
    next_button = driver.find_element(By.CLASS_NAME, "next")
    next_button.click()
    time.sleep(3)
    scrape_data()

df=pd.DataFrame({"Quote":Quote_title,'Author':Author_title,'Type_of_quotes':Type_of_quotes_title})
df
```

# Q7

```python
driver=webdriver.Chrome()
driver.get("https://www.jagranjosh.com/general-knowledge/list-ofall-prime-ministers-of-india-1473165149-1")

indexes = []
Name_title=[]
Born_Dead_title=[]
Term_of_office_title=[]
Remarks_title=[]

all_rows = driver.find_elements(By.CSS_SELECTOR, ".TableData tr td")

for index, item in enumerate(all_rows):
    if(index < 95):
        if((index + 1) % 5 == 1 ):
                indexes.append(item.text)
        elif((index + 1) % 5 == 2 ):
            Name_title.append(item.text)
        elif((index + 1) % 5 == 3):
            Born_Dead_title.append(item.text)
        elif((index + 1) % 5 == 4):
            Term_of_office_title.append(item.text)
        elif((index + 1) % 5 == 0):
            Remarks_title.append(item.text)
    
df=pd.DataFrame({"Name":Name_title,'Born_Dead':Born_Dead_title,'Term_of_office':Term_of_office_title,'Remarks':Remarks_title})
df
    
```

# Q8

```python
car_name_title=[]
Price_title=[]

driver=webdriver.Chrome()
driver.get("https://www.motor1.com/")

body_elem = driver.find_element(By.TAG_NAME, 'body')
body_elem.send_keys(Keys.ESCAPE)


search_box = driver.find_element(By.CSS_SELECTOR, '.m1-search-panel-input.m1-search-form-text')
search_box.send_keys('50 most expensive cars')


search_btn = driver.find_element(By.CSS_SELECTOR, '.m1-search-panel-button.m1-search-form-button-animate.icon-search-svg')
search_btn.click()

time.sleep(1)

first_link = driver.find_element(By.XPATH, '//div[@class="text-box"]/h3/a')
first_link.click()


car_names = driver.find_elements(By.CSS_SELECTOR, 'h3.subheader')
for index, name in enumerate(car_names):
    if(index < 50):
        car_name_title.append(name.get_attribute('innerText'))


car_prices = driver.find_elements(By.CSS_SELECTOR, 'p> strong')
for price in car_prices:
    Price_title.append(price.get_attribute('innerText'))

df=pd.DataFrame({"car_name":car_name_title,'Price':Price_title})
df

```
