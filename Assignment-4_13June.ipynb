# Q2

```python
driver=webdriver.Chrome()
driver.get("https://www.bcci.tv/")

fixture=driver.find_element(By.XPATH, "//div[@id='imw-international-men']//a[@class='border-0 '][normalize-space()='Fixtures & Results']")
fixture.click()

time.sleep(50)

Series_Title=[]
Place_Location=[]
Date_Required=[]
Time_Mentioned=[]

all_matches = driver.find_elements(By.XPATH, "//div[@class='match-info']")

for m in all_matches:
    series_elem = m.find_element(By.XPATH, ".//h5")
    Series_Title.append(series_elem.text)
    
    place_elem = m.find_element(By.XPATH, ".//div[@class='match-venue ng-scope']")
    Place_Location.append(place_elem.text)
    
    
    date_elem = m.find_element(By.XPATH, ".//div[@class='match-dates ng-binding']")
    Date_Required.append(date_elem.text)
    
    time_elem = m.find_element(By.XPATH, ".//div[@class='match-time no-margin ng-binding']")
    Time_Mentioned.append(time_elem.text)

df=pd.DataFrame({"Series":Series_Title,'Place':Place_Location,'Date':Date_Required,'Time':Time_Mentioned})
df
```

# Q3

```python
driver=webdriver.Chrome()
driver.get("https://www.statisticstimes.com/economy/india-statistics.php")

link=driver.find_element(By.XPATH,"//a[normalize-space()='» GDP of Indian states']")
link.click()

Rank_Title=[]
State_Title=[]
GSDP_22_23=[]
GSDP_21_22=[]
Share_18_19_Required=[]
GDP_Mentioned=[]

rank_tags=driver.find_elements(By.XPATH,"//td[@class='data1']")
for i in rank_tags:
    rank=i.text
    Rank_Title.append(rank)
mid = len(Rank_Title) // 2
Rank_Title = Rank_Title[:mid]
Rank_Title.pop()

State_tags=driver.find_elements(By.XPATH,"//td[@class='name']")
for i in State_tags:
    State=i.text
    State_Title.append(State)
mid_S = len(State_Title) // 2
State_Title = State_Title[:mid_S]
State_Title.pop()

GSDP1_tags=driver.find_elements(By.XPATH,"//*[@id='table_id']/tbody/tr/td[3]")
for i in GSDP1_tags:
    GSDP1=i.text
    GSDP_22_23.append(GSDP1)
    
GSDP2_tags=driver.find_elements(By.XPATH,"//*[@id='table_id']/tbody/tr/td[4]")
for i in GSDP2_tags:
    GSDP2=i.text
    GSDP_21_22.append(GSDP2)
    
Share_18_19_tags=driver.find_elements(By.XPATH,"//*[@id='table_id']/tbody/tr/td[5]")
for i in Share_18_19_tags:
    Share_18_19=i.text
    Share_18_19_Required.append(Share_18_19)

GDP_tags=driver.find_elements(By.XPATH,"//*[@id='table_id']/tbody/tr/td[6]")
for i in GDP_tags:
    GDP=i.text
    GDP_Mentioned.append(GDP)

df=pd.DataFrame({"Rank":Rank_Title,'State':State_Title,'GSDP(22-23)':GSDP_22_23,'GSDP(21-22)':GSDP_21_22,'Share(18-19)':Share_18_19_Required,"GDP":GDP_Mentioned})
df
```

# Q4

```python
driver=webdriver.Chrome()
driver.get("https://github.com/trending")

Repository_Title=[]
Repository_Description=[]
Contributors_Count=[]
Language_Used=[]
langs=[]
stars=[]
    
all_repos = driver.find_elements(By.XPATH, "//article[@class='Box-row']")
for repo in all_repos:
    try:
        lang = repo.find_element(By.XPATH,".//span[@itemprop='programmingLanguage']")
        langs.append(lang.text)
    except NoSuchElementException as e:
        langs.append('-')
    star_elem = repo.find_element(By.XPATH,".//a[@class='Link Link--muted d-inline-block mr-3']")
    stars.append(star_elem.text)
    title_elem = repo.find_element(By.XPATH,".//h2[@class='h3 lh-condensed']/a")
    Repository_Title.append(title_elem.text)
Language_Used = langs

Description_tags=driver.find_elements(By.XPATH,"//p[@class='col-9 color-fg-muted my-1 pr-4']")
for i in Description_tags:
    Description=i.text
    Repository_Description.append(Description)


df=pd.DataFrame({"Repository_Title":Repository_Title,'Repository_Description':Repository_Description,'Total_Stars':stars,'Language_Used':Language_Used})
df
```

# Q7

```python
driver=webdriver.Chrome()
driver.get("https://www.imdb.com/list/ls512407256/")

Name_Title=[]
Year_span=[]
Genre_name=[]
Run_time=[]
Ratings=[]
Votes_Count=[]

all_shows = driver.find_elements(By.XPATH, "//div[@class='sc-74bf520e-2 cVXqoq']")
for s in all_shows:
    title_elem = s.find_element(By.XPATH, ".//h3[@class='ipc-title__text']")
    Name_Title.append(title_elem.text.split('.')[1])
    
    year_elem = s.find_elements(By.XPATH, ".//div[@class='sc-b189961a-7 feoqjK dli-title-metadata']/span")[0]
    Year_span.append(year_elem.text)
    
    runtime_elem = s.find_elements(By.XPATH, ".//div[@class='sc-b189961a-7 feoqjK dli-title-metadata']/span")[1]
    Run_time.append(runtime_elem.text)
    
    rating_elem = s.find_element(By.XPATH, ".//div[@class='sc-e2dbc1a3-0 ajrIH sc-b189961a-2 fkPBP dli-ratings-container']/span")
    Ratings.append(rating_elem.text.split('\n')[0])
    Votes_Count.append(rating_elem.text.split('\n')[1].replace('(', '').replace(')', ''))


df=pd.DataFrame({"Name":Name_Title,'Year_span':Year_span,'Run_time':Run_time,'Ratings':Ratings,"Votes":Votes_Count})
df
```

# Q8

```python
driver=webdriver.Chrome()
driver.get("https://archive.ics.uci.edu/")

time.sleep(3)

view_dataset_btn = driver.find_element(By.XPATH, "//a[@class='btn-primary btn']")
view_dataset_btn.click()

time.sleep(5)

expand_btn = driver.find_element(By.XPATH, "//label[@class='btn-primary btn-sm btn flex gap-2 rounded-full']")
expand_btn.click()

Dataset_Name=[]
Data_Type=[]
Task_Name=[]
Attribute_Types=[]
No_of_instances=[]
No_of_attributes =[]
Year_mentioned=[]

Dataset_tags=driver.find_elements(By.XPATH,"//a[@class='link-hover link text-xl font-semibold']")
for i in Dataset_tags:
    Dataset=i.text
    Dataset_Name.append(Dataset)

all_datasets_container = driver.find_elements(By.XPATH, "//div[@class='my-2 hidden gap-4 md:grid grid-cols-12']")
for c in all_datasets_container:
    type_name = c.find_elements(By.XPATH, ".//span[@class='truncate']")[1]
    Data_Type.append(type_name.text)
    
    instance_elem = c.find_elements(By.XPATH, ".//span[@class='truncate']")[2]
    No_of_instances.append(instance_elem.text.split(' ')[0])
    
    feature_elem = c.find_elements(By.XPATH, ".//span[@class='truncate']")[3]
    No_of_attributes.append(feature_elem.text.split(' ')[0])
    
    task_elem = c.find_elements(By.XPATH, ".//span[@class='truncate']")[0]
    Task_Name.append(task_elem.text)
    
    
date_tables = driver.find_elements(By.XPATH, "//table[@class='col-span-full my-2 table sm:col-start-2']")
for t in date_tables:
    td_elem = t.find_elements(By.XPATH, ".//tbody/tr/td")[2]
    Year_mentioned.append(td_elem.text.split('/')[2])
    
    feature_type_elem = t.find_elements(By.XPATH, ".//tbody/tr/td")[1]
    Attribute_Types.append(feature_type_elem.text)


df=pd.DataFrame({"Dataset_Name":Dataset_Name,'Data_Type':Data_Type,'Task':Task_Name,'Attribute':Attribute_Types,'No_of_instances':No_of_instances,"No_of_attributes":No_of_attributes,"Year":Year_mentioned})
df
```
