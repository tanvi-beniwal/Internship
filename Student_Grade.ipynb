Making the data ready for applying various models but diving it into dependent and independent variable. Spliting it in Test and train.

```python
X = df.drop(['CGPA'], axis =1)
Y = df['CGPA']
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
X_scale = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import cross_val_score,GridSearchCV
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,test_size=.30,random_state=65)
print('Training size:',X_train.shape)
print('Training target size:',Y_train.shape)
print('Test size:',X_test.shape)
print('Test target size:',Y_test.shape)
```

```python
lm=LinearRegression()
lm.fit(X_train,Y_train)
lm.score(X_test,Y_test)
pred1=lm.predict(X_train)
pred=lm.predict(X_test)
print("Mean absolute Error",mean_absolute_error(Y_test,pred))
print("Mean squared error",mean_squared_error(Y_test,pred))
print("r2 Score on test data",r2_score(Y_test,pred)*100)  
print("r2 Score on training data",r2_score(Y_train,pred1)*100)
```

```python
sv=SVR()
sv.fit(X_train,Y_train)
sv.score(X_test,Y_test)
pred1=sv.predict(X_train)
pred=sv.predict(X_test)
print("Mean absolute Error",mean_absolute_error(Y_test,pred))
print("Mean squared error",mean_squared_error(Y_test,pred))
print("r2 Score on test data",r2_score(Y_test,pred)*100)   
print("r2 Score on training data",r2_score(Y_train,pred1)*100)
```

```python
ar=AdaBoostRegressor()
ar.fit(X_train,Y_train)
ar.score(X_test,Y_test)
pred1=ar.predict(X_train)
pred=ar.predict(X_test)
print("Mean absolute Error",mean_absolute_error(Y_test,pred))
print("Mean squared error",mean_squared_error(Y_test,pred))
print("r2 Score on test data",r2_score(Y_test,pred)*100) 
print("r2 Score on training data",r2_score(Y_train,pred1)*100)
```

```python
rf=RandomForestRegressor()
rf.fit(X_train,Y_train)
rf.score(X_test,Y_test)
pred1=rf.predict(X_train)
pred=rf.predict(X_test)
print("Mean absolute Error",mean_absolute_error(Y_test,pred))
print("Mean squared error",mean_squared_error(Y_test,pred))
print("r2 Score on test data",r2_score(Y_test,pred)*100)  
print("r2 Score on training data",r2_score(Y_train,pred1)*100)
```

All the models above shows R-squared Score on test and train data also Mean absolute error and Mean squared error can be seen. From the above models the best suited model for this dataset is Random Forest Regressor model with 98.9% accuracy but still will be further to do cross validation on all the models to select the best suited model the given dataset.


Using cross_val_score to perform cross-validation. It evaluates the model's performance using k-fold cross validation.

```python
cv=cross_val_score(rf,X,Y,scoring="r2")
print(cv)
print(cv.mean())
print(cv.std())
```

```python
cv1=cross_val_score(sv,X,Y)
print(cv1)
print(cv1.mean())
print(cv1.std())
```

```python
cv2=cross_val_score(lm,X,Y)
print(cv2)
print(cv2.mean())
print((cv2.std()))
```

```python
cv3=cross_val_score(ar,X,Y)
print(cv3)
print(cv3.mean())
print((cv3.std()))
```
