OVERVIEW OF PROJECT
Health insurance charges have risen dramatically over the past decade in response to the rising cost of health care services and are determined by a many factors. The cahrges can be because of many factors like  age, sex, bmi, number of children, smoker,sex and region.
The purpose of this project is to determine how the above mentioned factors helps in predict health insurance charges by performing exploratory data analysis and predictive modeling on the Health Insurance dataset. In this project following libaries are used for performing various functions on the data set:-
Numpy, Pandas, Sci-kit learn,matplotlib.pyplot and seaborn.

ABOUT THE DATASET
1. age: age of beneficiary
2. sex: insurance individuals gender, female, male
3. bmi: Body mass index, providing an understanding of body, weights that are relatively high or low relative to            height.
4. children: Number of children covered by health insurance.
5. smoker: whether the individuak somke or not.
6. region: the beneficiary's residential area  northeast, southeast, southwest, northwest
7. charges: Individual medical charges need to be pay.
