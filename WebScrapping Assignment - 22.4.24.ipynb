```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#Question_1
movie_names = []
release_years = []
ratings = []
page1=requests.get("https://www.imdb.com/list/ls056092300/")
movie_soup=BeautifulSoup(page1.content)
movie_list = movie_soup.find_all("h3", class_="lister-item-header")
for c in movie_list:
    name = c.contents[3].text
    movie_names.append(name)
for y in movie_soup.find_all("span",class_="lister-item-year text-muted unbold"):
    release_years.append(y.text.replace('(', '').replace(')', ''))
for r in movie_soup.find_all("div",class_="ipl-rating-star small"):
    ratings.append(r.contents[3].text)
    
    
df=pd.DataFrame({"Name":movie_names," Rating":ratings,"Year of release":release_years})
df
```

```python
#Question 2

product_names=[]
price=[]
discounts=[]

page2=requests.get("https://peachmode.com/pages/search?q=bags")
soup=BeautifulSoup(page2.content)

for n in soup.find_all("div",class_="product-title"):
    product_names.append(n.text)
for p in soup.find_all("span",class_="price st-money money done"):
    price.append(p.text)
for d in soup.find_all("span",class_="discounted_price st-money money done"):
    discounts.append(d.text) 

df=pd.DataFrame({"Name":product_names,"Price":price,"Discount":discounts})
df
```

```python
#Question 3
# a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

top_10_teams=[]
matches=[]
points=[]
ratings=[]

page3_teams=requests.get("https://www.icc-cricket.com/rankings/team-rankings/mens/odi")
soup=BeautifulSoup(page3_teams.content)

for tt in soup.find_all("div",class_="si-team-name"):
    top_10_teams.append(tt.contents[1].text)
for m in soup.find_all("div",class_="si-table-data si-matches"):
    matches.append(m.span.text)
for p in soup.find_all("div",class_="si-table-data si-pts"):
    points.append(p.span.text)
for r in soup.find_all("div",class_="si-table-data si-rating"):
    ratings.append(r.span.text)
    
df = pd.DataFrame({"Top_10_Teams":top_10_teams,"Matches":matches,"Points":points,"Ratings":ratings})
df
```

```python
#Question 3
# b) Top 10 ODI Batsmen along with the records of their team and rating.

top_10_batsmen=[]
batsmen_team=[]
batsmen_rating=[]

page3=requests.get("https://www.icc-cricket.com/rankings/batting/mens/odi")
soup=BeautifulSoup(page3.content)

for tb in soup.find_all("a",class_="si-player-name-wrap"):
    top_10_batsmen.append(tb.contents[0].text + tb.contents[1].text)
for bt in soup.find_all("h3",class_="si-team-name"):
    batsmen_team.append(bt.contents[1].text)
for br in soup.find_all("div",class_="si-table-data si-rating"):
    batsmen_rating.append(br.span.text)

df = pd.DataFrame({"Top_10_Batsmen":top_10_batsmen,"Batsmen_Team":batsmen_team,"Batsmen_Rating":batsmen_rating})
df
```

```python
#Question 3
#c) Top 10 ODI bowlers along with the records of their team and rating

top_10_bowlers=[]
team=[]
rating=[]

page3=requests.get("https://www.icc-cricket.com/rankings/bowling/mens/odi")
soup=BeautifulSoup(page3.content)

for t in soup.find_all("h3",class_="si-player-name"):
    first_name = t.a.contents[0].text
    last_name = t.a.contents[1].text
    top_10_bowlers.append(first_name + last_name)
for m in soup.find_all("h3",class_="si-team-name"):
    team.append(m.contents[1].text)
for r in soup.find_all("div",class_="si-table-data si-best"):
    rating.append(r.span.text) 

df=pd.DataFrame({"Top_10_Bowlers":top_10_bowlers,"team":team,"Rating":rating})
df
```

```python
# Question 4
# Write a python program to scrape details of all the posts from https://www.patreon.com/coreyms .Scrape the
# heading, date, content and the likes for the video from the link for the youtube video from the post.
heading=[]
date=[]
contents=[]
likes=[]

page4=requests.get("https://www.patreon.com/coreyms")
soup=BeautifulSoup(page4.content)

for h in soup.find_all("div",class_="sc-bBHxTw iloeMK"):
    heading.append(h.text)
for d in soup.find_all("p",class_="sc-bqiRlB etUZPh sc-lgu5zg-1 bDqjCN"):
    date.append(d.span.button.span.text)
for c in soup.find_all('div', class_='sc-cfnzm4-0 kJujbw'):
    contents.append(c.p.text)
for l in soup.find_all("div",class_="sc-jrQzAO knZGTZ"):
    likes.append(l.span.text) 

df=pd.DataFrame({"Heading":heading,"Date":date,"Likes":likes, "Content": contents})
df
```

```python
#Question 5 Indira Nagar

house_title=[]
location=[]
area=[]
EMI=[]
price=[]

page5=requests.get("https://www.nobroker.in/flats-for-sale-in-indira_nagar_bangalore")
soup=BeautifulSoup(page5.content)

for h in soup.find_all("h2",class_="heading-6 flex items-center font-semi-bold m-0"):
    house_title.append(h.a.text)
for l in soup.find_all("div",class_="mt-0.5p overflow-hidden overflow-ellipsis whitespace-nowrap max-w-70 text-gray-light leading-4 po:mb-0.1p po:max-w-95"):
    location.append(l.text)
for a in soup.find_all("div",class_="flex flex-col w-33pe items-center tp:w-half po:w-full"):
    area.append(a.div.div.text) 
for e in soup.find_all("div",class_="p-1.5p flex border-b border-b-solid border-cardbordercolor tp:py-1p tp:px-1.5p tp:border-b-0"):
    EMI.append(e.contents[2].div.contents[1].text)
    price.append(e.contents[1].div.span.text)

df=pd.DataFrame({"House_Title":house_title,"Location":location,"Area":area,"EMI":EMI,"Price":price})
df
```

```python
#Question 6

product_name=[]
price=[]
images=[]

page6=requests.get("https://www.bewakoof.com/bestseller?sort=popular")
soup=BeautifulSoup(page6.content)

for pn in soup.find_all("div",class_="productNaming bkf-ellipsis"):
    product_name.append(pn.h2.text)
for p in soup.find_all("div",class_="discountedPriceText clr-p-black false"):
    price.append(p.contents[1].text)
for i in soup.find_all("img",class_="productImgTag"):
    images.append(i.get('src'))

df=pd.DataFrame({"Product_Name":product_name,"Price":price,"Image":images})
df
```

```python
#Question 7
headings=[]
dates=[]
news_links=[]

page7=requests.get("https://www.cnbc.com/world/?region=world")
soup=BeautifulSoup(page7.content)

for h in soup.find_all("div",class_="LatestNews-headlineWrapper"):
    if(h.contents[1].get('href') and h.contents[1].get('href').find('cnbc') > -1):
        headings.append(h.contents[1].text)
        
        news_link = h.contents[1].get('href')
        news_links.append(news_link)
        dates.append(h.span.time.text)

df=pd.DataFrame({"Headings":headings,"News_link":news_links, "Date": dates})
df

```

```python
#Question 8
Paper_title=[]
date=[]
Author=[]

page8=requests.get("https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/")
soup=BeautifulSoup(page8.content)
for p in soup.find_all("h2",class_="h5 article-title"):
    Paper_title.append(p.a['href'])
for d in soup.find_all("p",class_="article-date"):
    date.append(d.text)
for n in soup.find_all("p",class_="article-authors"):
    Author.append(n.text) 

df=pd.DataFrame({"Paper_title":Paper_title,"Date":date,"Author":Author})
df
```
